```python
check=0
count=0


with open("messages.ndjson.zst", 'rb') as fh:
    dctx = zstd.ZstdDecompressor()
    with dctx.stream_reader(fh) as reader:
        
        
        
        previous_line = ""
        while True:
            chunk = reader.read(2**30) # 26 is  default, 31 is max
            if not chunk:
                break

            string_data = chunk.decode('utf-8',errors='ignore')
            lines = string_data.split("\n")
            
            check=check+1
            print("--check: ",check)
            
            list=[]
            
            for i, line in enumerate(lines[:-1]):
                if i == 0:
                    line = previous_line + line
                object = json.loads(line)
                
                #print(object)
                #print("====================")
                
                keys = ['message','date','from_id','id','reply_to_msg_id']
                
                try:
                    filtered_object = dict((k, object[k]) for k in keys)
                    list.append(filtered_object)
                except Exception:
                    pass

            df = pd.DataFrame(list)
        
            ##data cleanning
            #df.drop(df.loc[df['message'] == "\\"].index, inplace=True)
            df.drop(df.loc[df['message'].str.len() <= 2].index, inplace=True)
            df.drop(df.loc[df['message'] == '^'].index, inplace=True)
            df.drop(df.loc[df['message'] == ''].index, inplace=True)
            df['message'] = df['message'].str.replace('|',',').replace(r'\n', '', regex=True)
            df['date'] = df['date'].str.slice(0,10)
            df['date']= pd.to_datetime(df['date'], errors='coerce', format='%Y-%m-%d').dt.strftime('%m/%d/%Y')
            
            ###detect language
            #i=0
            #df3 = pd.DataFrame()
            #while i < len(df):
            #    text = client.detect(df.iloc[i]["message"]).language
          #
            #    if text == "en":    
            #        df2 = {"message": text, "date": df.iloc[i]["date"]}
            #        df3 = df3.append(df2, ignore_index = True) 
            #        print(df3.shape)
            #    i=i+1
            
            df.to_csv(f'messages_SBS_{check}.csv',sep = '|', index=False)
                
            #n = 100000
            #
            #for g, df in df.groupby(np.arange(len(df)) // n):
            #    df.to_csv(f'messages_SBS_{count}.csv',sep = '|', index=False)
            #    print(g)
            #    count = count + 1
                
                
            #def save_file_part(df, size_threshold, save_path, part_number=0):
            #    file_size = df.memory_usage(index=True, deep=False).sum() / 1038336
            #    num_records = len(df)
            #
            #    if file_size > size_threshold:
            #        records_to_split_off = int(num_records * size_threshold // file_size)
            #        df_to_save = df.head(records_to_split_off)
            #        df_to_save.to_csv(f'messages_SBS_{check}_{part_number}.csv',sep='|', index=False, encoding='utf-8')
            #        save_file_part(df.tail(num_records-records_to_split_off), size_threshold, save_path, part_number=part_number+1)
            #
            #    else:
            #        df.to_csv(f'messages_SBS_{check}_{part_number}.csv', sep='|', index=False, encoding='utf-8')
            #

            #save_file_part(df, 3, save_path='F:\\DataCleaning\\Telegram Dataset\\Ukrainetest{}.csv') 

                

            previous_line = lines[-1]
```

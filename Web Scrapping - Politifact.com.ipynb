```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("https://www.politifact.com/")

totalContent = response.content

soup = BeautifulSoup(totalContent, 'html.parser') # Let bs4 ingest this raw html and parse it. There are different types of html parsers you can use.
```

### Create a empty dataframe and set column

```python
df = pd.DataFrame()
df['text'] = df.mean(0)
df['result'] = df.mean(1)
```

```python
tag=soup.find_all(class_='m-statement__body')
for i in tag: 
    try:
        df_add = {"text": str(i.find("a").get_text()), "result": i.img['alt']} # only need the 'alt' information from image
        df = df.append(df_add, ignore_index = True)
    except Exception:
        pass
```

### Data cleaning

```python
df["text"] = df["text"].str.slice(1,-1)
```
